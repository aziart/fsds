### Flight Price Prediction (EDA, Feature Engineering)

#### Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
trn = pd.read_excel("data/flight_price/train.xlsx")
tst = pd.read_excel("data/flight_price/test.xlsx")
display(trn.head(), trn.shape)
display(tst.head(), tst.shape)
```

```python
trn.info()
```

```python
trn['Date_of_Journey'].str.split('/').str[0]
```

```python
trn['Journey_Date'] = trn['Date_of_Journey'].str.split('/').str[0]
trn['Journey_Month'] = trn['Date_of_Journey'].str.split('/').str[1]
trn['Journey_Year'] = trn['Date_of_Journey'].str.split('/').str[2]
display(trn.head())
trn.info()
```

```python
trn['Journey_Date'] = trn['Journey_Date'].astype(int)
trn['Journey_Month'] = trn['Journey_Month'].astype(int)
trn['Journey_Year'] = trn['Journey_Year'].astype(int)
display(trn.head())
trn[['Journey_Date', 'Journey_Month', 'Journey_Year']].info()
```

```python
trn.drop('Date_of_Journey', axis=1, inplace=True)
display(trn.columns)
```

```python
trn.head(2) # note: arrival time
```

```python
trn['Arrival_Time'] = trn['Arrival_Time'].str.split(' ').str[0]
# trn['Arrival_Time'] = trn['Arrival_Time'].apply(lambda x: x.split(' ')[0])
trn['Arrival_Hour'] = trn['Arrival_Time'].str.split(':').str[0]
trn['Arrival_Minute'] = trn['Arrival_Time'].str.split(':').str[1]
display(trn.head(2))
```

```python
display(trn[['Arrival_Hour', 'Arrival_Minute']].dtypes)
trn['Arrival_Hour'] = trn['Arrival_Hour'].astype(int)
trn['Arrival_Minute'] = trn['Arrival_Minute'].astype(int)
display(trn[['Arrival_Hour', 'Arrival_Minute']].dtypes)
```

```python
trn['Dep_Hour'] = trn['Dep_Time'].str.split(':').str[0]
trn['Dep_Minute'] = trn['Dep_Time'].str.split(':').str[1]
display(trn.head(2))
```

```python
display(trn[['Dep_Hour', 'Dep_Minute']].dtypes)
trn['Dep_Hour'] = trn['Dep_Hour'].astype(int)
trn['Dep_Minute'] = trn['Dep_Minute'].astype(int)
display(trn[['Dep_Hour', 'Dep_Minute']].dtypes)
```
