```python
import plotly
import plotly.express as px
# import plotly.io as pio
# pio.renderers
import cufflinks as cf
cf.go_offline() #will make cufflinks offline
cf.set_config_file(offline=False, world_readable=True)
```

```python
iris
```

```python
"""
will not be displayed on github, because github performs a static render of the notebooks and it
doesn't include the embedded HTML/JavaScript that makes up a plotly graph.
Solution is to paste the link to your GitHub notebook into http://nbviewer.jupyter.org/, which will present a rich view of the notebook.
"""
l6_15 = cf.Layout(
    height=600,
    width=1500,
    margin=dict(l=35, r=35, t=35, b=35),
    paper_bgcolor="LightSteelBlue"
)

fig = iris.iplot(asFigure=True,
layout=l6_15)
fig.show()
fig.write_html('0326 iplots/0326v - 40.html')
fig.write_image('0326 iplots/0326v - 40.png')
```

```python
fig = iris.iplot(x = 'SepalLengthCm',
                y = 'PetalLengthCm',
                mode = 'markers',
                kind = 'scatter',
                asFigure=True,
                layout=l6_15)
fig.show()
fig.write_html('0326 iplots/0326v - 41.html')
fig.write_image('0326 iplots/0326v - 41.png')
```

```python
fig = iris.iplot(mode = 'markers',
                kind = 'scatter',
                asFigure=True,
                layout=l6_15)
fig.show()
fig.write_html('0326 iplots/0326v - 42.html')
fig.write_image('0326 iplots/0326v - 42.png')
```

```python
fig = iris.iplot(mode = 'markers',
                kind = 'scatter',
                asFigure=True,
                layout=l6_15,
                size=7)
fig.show()
fig.write_html('0326 iplots/0326v - 43.html')
fig.write_image('0326 iplots/0326v - 43.png')
```

```python
fig = iris.iplot(x = 'SepalLengthCm',
                y = 'PetalLengthCm',
                size = 'SepalLengthCm',
                kind = 'bubble',
                asFigure=True,
                layout=l6_15)
fig.show()
fig.write_html('0326 iplots/0326v - 44.html')
fig.write_image('0326 iplots/0326v - 44.png')
```

```python
iris.scatter_matrix()
import plotly.express as px

fig = px.scatter_matrix(iris, height=1000, width=1000)
fig.show()
fig.write_html('0326 iplots/0326v - 45alt.html')
fig.write_image('0326 iplots/0326v - 45alt.png')
```

```python
l12_10 = cf.Layout(height=1200,
                    width=1000,
                    margin=dict(l=20, r=20, t=20, b=20),
                    paper_bgcolor="LightSteelBlue")

fig = iris.iplot(kind = 'scatter3d',
                x = 'SepalLengthCm',
                y = 'SepalWidthCm',
                z = 'PetalLengthCm',
                asFigure=True,
                layout=l12_10)

fig.show()
fig.write_html('0326 iplots/0326v - 47.html')
fig.write_image('0326 iplots/0326v - 47.png')
```

```python
l9_12 = cf.Layout(height=900, width=1200,
                margin=dict(l=20, r=20, t=20, b=20),
                paper_bgcolor="LightSteelBlue")
fig = iris.head(10).iplot(kind = 'scatter3d',
                            x = 'SepalLengthCm',
                            y = 'SepalWidthCm',
                            z = 'PetalLengthCm',
                            asFigure=True,
                            layout=l9_12)

fig.show()
fig.write_html('0326 iplots/0326v - 48.html')
fig.write_image('0326 iplots/0326v - 48.png')
```

```python
fig = iris.iplot(kind = 'bubble3d',
x = 'SepalLengthCm',
y = 'SepalWidthCm',
z = 'PetalLengthCm',
size='PetalWidthCm',
asFigure=True,
layout=l12_10)

fig.show()
fig.write_html('0326 iplots/0326v - 49.html')
fig.write_image('0326 iplots/0326v - 49.png')
```
