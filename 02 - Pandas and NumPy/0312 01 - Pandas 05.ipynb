```python
def first_o(x):
    if x.lower() == 'o':
        return 'YES'
    else:
        return 'NO'
```

```python
df['col6'] = df['col5'].apply(lambda x: first_o(x))
df
```

```python
df['col6'] = df['col5'].apply(first_o)
df
```
